```python
from datasets import load_dataset
import os

# Tải tập dữ liệu
ds = load_dataset("zh-plus/tiny-imagenet")

# Tạo thư mục lưu trữ
val_dir = "/mnt/mdata/new/imagenet/val"
os.makedirs(val_dir, exist_ok=True)

# Lưu dữ liệu từ split 'valid'
for i, example in enumerate(ds['valid']):  # 'valid' chứa tập validation
    label = example['label']  # Nhãn lớp (dạng số)

    # Tạo thư mục cho mỗi nhãn
    class_dir = os.path.join(val_dir, str(label))
    os.makedirs(class_dir, exist_ok=True)

    # Đường dẫn ảnh
    img_path = os.path.join(class_dir, f"{label}_{i}.jpg")

    # Lưu ảnh
    example['image'].save(img_path, format='JPEG')

print(f"Dữ liệu đã được lưu vào {val_dir}")
```
